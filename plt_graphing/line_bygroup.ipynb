```python
# Import
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style("whitegrid")
```

```python
# Import .csv file
datafile = pd.read_csv("/Users/noahwaller/Documents/3cohort-GIMME PAPER/csv_for-python/mihyst_visqst_unpl.csv")
datafile.head()
```

```python
# Define list flattener function (useful for simplifying matplotlib)
# no longer necessary but kept here for the future
def flatten(xss):
    return [x for xs in xss for x in xs]
```

```python
# Set font size, type, and style for all graphs
font = {'family' : 'arial',
        'weight' : 'normal',
        'size'   : 16}

plt.rc('font', **font)
```

```python
# Make line graph of visual unpleasantness ratings across 6 brightness levels

plt.figure(figsize=(8,5)) # set fig size

x = [1, 2, 3, 4, 5, 6] # brightness levels
y = [datafile['vis01_unpl_avg'].mean(), datafile['vis02_unpl_avg'].mean(), datafile['vis03_unpl_avg'].mean(),
     datafile['vis04_unpl_avg'].mean(), datafile['vis05_unpl_avg'].mean(), datafile['vis06_unpl_avg'].mean()] # calc mean

yerror = [datafile['vis01_unpl_avg'].std(), datafile['vis02_unpl_avg'].std(), datafile['vis03_unpl_avg'].std(), 
          datafile['vis04_unpl_avg'].std(), datafile['vis05_unpl_avg'].std(), datafile['vis06_unpl_avg'].std()] # calc SD

plt.plot(x, y, marker='o', color='b', label='MIHYST CPP') # plot x and y, set custom details

plt.errorbar(x, y, yerr=yerror) # add error bars

plt.ylim([0, 100]) # set y axis limit

plt.ylabel('Visual Unpleasantness (0-100)') # ylabel
plt.xlabel('Brightness Level') # xlabel
plt.legend() # plot legend based on labels

plt.show() # show plot
```
